```python
data_df = data[['date','location','total_vaccinations','total_vaccinations_per_hundred','total_distributed',
                'distributed_per_hundred','people_vaccinated', 'people_vaccinated_per_hundred','people_fully_vaccinated',
                'people_fully_vaccinated_per_hundred','daily_vaccinations','daily_vaccinations_per_million',
                'total_boosters','total_boosters_per_hundred']].dropna()
data_df
```

```python
data_df.head()
```

```python
data_df.info()
```

```python
data_df['date'] = pd.to_datetime(data_df['date'])
```

```python
data_df.info()
```

```python
year = 2022
new_df_year = data_df[data_df['date'].dt.year == year]
new_df_year
```

```python
new_df_year.fillna(0, inplace=True)
```

```python
new_df_year['location'].unique()
```

```python
# drop locations that are not states

drop_locations = ['Marshall Islands','Indian Health Svc','United States','American Samoa','District of Columbia','Federated States of Micronesia','Northern Mariana Islands',
            'Virgin Islands','Republic of Palau','Puerto Rico','Guam']
new_df_year.drop(new_df_year[new_df_year['location'].isin(drop_locations)].index, inplace=True)
new_df_year.shape
```

```python
location_df= new_df_year['location'].unique()
location_df
```

```python
new_data_df_grouped = new_df_year.groupby(['location']).sum().reset_index()
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
new_data_df_grouped.head()
```

```python
new_data_df_grouped.to_csv("data/Covid_vaccination.csv", header=True)
```
